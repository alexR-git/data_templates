```python
def donwload_from_url(url, filename, sep=",", delim="\n"): # Separador por defecto: ",""  y delimitador por defecto: salto de línea
    
    import pandas as pd
    import openpyxl
    # Importado de urllib3 y petición al servidor
    import urllib3
    
    http = urllib3.PoolManager()         # Objeto PoolManager para poder enviar requests/peticiones. Este objeto maneja todos los detalles del pool de conexiones por ti.
    r = http.request('GET', url)         # Crea una petición GET a la url indicada y guarda la respuesta del servidor
    #response_medals.auto_close = False  # By default HTTP responses are closed after reading all bytes, this disables that behavior
    #response_medals.status              # Estado del servidor
    response = r.data                    # Datos del dataset

    # response es una lista de strings codificados como strings binarios. Decodificación de estos en UTF-8
    str_data = response.decode('utf-8')
    str_data

    # El dataset ahora tiene pinta de una única cadena de texto donde las celdas están separadas por comas y las filas están delimitadas por \n,
    # por esto, spliteo por \n y queda el dataset separado por listas de strings separados por comas
    lines = str_data.split(delim)
    
#     Print de control: dataset en listas de strings separadas por ,
#     print(f"El dataset en listas de strings separadas por comas es: \n\n{lines}")

    # En la primera fila están las cabeceras de las columnas, así que la extraemos:
    col_names = lines[0].split(sep) # spliteo por comas porque cada fila es una lista de strings separada de la siguiente por comas
    
#     # print de control: nombres de las columnas
#     print(f"Nombres de las columnas: {col_names} \n")
    
    num_cols = len(col_names)       # número de columnas

    # Genera un diccionario vacío para el dataset
    dict_urllib3 = {}

    # Y se generan las keys
    for key in col_names:
        dict_urllib3[key] = []

    # Rellenado del diccionario
    contador = 0

    for line in lines: 
        if (contador > 0):                   # if para que no lea en values la primera línea: lleva las cabeceras de las columnas
            values = line.strip().split(sep) # cada fila es una lista de strings separada por comas; se splitea y se va leyendo
            #print(values)
            for i in range(len(col_names)):  # Y cuando ya la fila está leída (values), se asignan sus celdas a las keys.
                dict_urllib3[col_names[i]].append(values[i])
        contador += 1

#     print de control: muestra el diccionario
#     print(f"Diccionario:\n", dict_urllib3)

    # Creación del dataframe
    data = pd.DataFrame(dict_urllib3)
    data.head()

    # Creación del archivo de salida sin formato de archivo
    mainpath = "[Ruta de salida para los datasets]"
    
#     Seteo local del nombre del archivo con el dataframe
#     filename = "[nombre del archivo que contiene el dataset]"
    path = mainpath + filename
    
#     print de control: ruta completa del archivo que contiene el dataset
#     print(path)

    # Formato de archivos de salida: json, csv y xls
    data.to_csv(path + ".csv")
    data.to_json(path + ".json")
    data.to_excel(path + ".xls", engine='openpyxl')
    
    print("Los ficheros se han guardado correctamente en: ", path)
    
    return data
```
